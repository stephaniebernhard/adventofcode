```python
# Day 18

def import_input():
    my_list = []
    with open('input18test.txt') as f:
        for line in f:
            cube = line.rstrip('\n').split(',')
            arr = []
            for el in cube:
                new_el = int(el)
                arr.append(new_el)
            my_list.append(arr)
    return my_list

inputlist = import_input()

def cubes_touch(cube1, cube2):
    z_touch = abs(cube1[0]-cube2[0]) == 0 and abs(cube1[1]-cube2[1]) == 0 and abs(cube1[2]-cube2[2]) == 1
    y_touch = abs(cube1[0]-cube2[0]) == 0 and abs(cube1[1]-cube2[1]) == 1 and abs(cube1[2]-cube2[2]) == 0
    x_touch = abs(cube1[0]-cube2[0]) == 1 and abs(cube1[1]-cube2[1]) == 0 and abs(cube1[2]-cube2[2]) == 0
    return x_touch or y_touch or z_touch

count = 0 

for i in range(len(inputlist)):
    for j in range(i+1,len(inputlist)):
        if cubes_touch(inputlist[i],inputlist[j]):
            count += 1
            
number_of_cubes = len(inputlist)
print("number_of_cubes: " + str(number_of_cubes))     

surfaces = number_of_cubes * 6
print("number of surfaces: " + str(surfaces))
print("number of touches: " + str(count))

surface_area = surfaces - 2*count
print("surface_area: " + str(surface_area))
```
